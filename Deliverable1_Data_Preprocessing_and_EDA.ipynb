```python
## Advanced Data Mining â€“ Deliverable 1: Data Collection, Cleaning, and Exploration

## This notebook walks through data loading, cleaning, and exploratory data analysis using a healthcare dataset.

## Step 1: Load and Preview the Dataset'

import pandas as pd
# Load dataset
df = pd.read_csv("healthcare_dataset.csv")
df.head()

## Step 2: Dataset Structure and Summary

# Shape of dataset
print("Shape:", df.shape)

# Information on dataset
df.info()

# Statistical summary
df.describe()

# Missing values
df.isnull().sum()

# Check duplicates
df.duplicated().sum()

## Step 3: Data Cleaning

# Fill numeric missing values with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill categorical missing values with mode
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
    
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
## Step 4: Exploratory Data Analysis (EDA)

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Histogram for Age
df['Age'].hist(bins=30)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Bar plot for Gender
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Boxplot Age by Gender
sns.boxplot(x='Gender', y='Age', data=df)
plt.title("Boxplot of Age by Gender")
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

## Step 5: Key Insights

# Age shows a skewed distribution with visible outliers.
# Gender is [balanced/skewed] in this dataset.
# Some attributes show strong correlations, which may be useful for modeling.
# Missing values were successfully imputed.
```
